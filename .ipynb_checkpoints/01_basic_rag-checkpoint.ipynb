## Retrieval-Augmented Generation (RAG) is an AI technique that combines the retrieval of relevant information from an external knowledge source with the text generation capabilities of a Large Language Model (LLM) to produce more accurate, up-to-date, and context-aware responses.

## Instead of relying only on the knowledge stored in the model's parameters during training, RAG first searches a knowledge base (such as documents, PDFs, databases, or web pages), retrieves the most relevant information, and then provides that information to the LLM as additional context before generating a response.

## Replacing Character-Based Splitting with BPE Tokenization

**Expected Changes:**

- Chunks are created based on **tokens** instead of **characters**.
- Each chunk has a more consistent size with respect to the model's token limit.
- Chunking becomes aligned with how LLMs and embedding models process text.
- The retrieval process can become more accurate, especially for long documents.
- This approach reduces the risk of exceeding token limits and is better suited for production RAG systems.


```python
import bs4  #  HTML parsing garna BeautifulSoup use garne

# from langchain_text_splitters import ( RecursiveCharacterTextSplitter)  #  Text lai chunk ma split garna

from langchain_community.document_loaders import (
    WebBaseLoader,
)  #  Web page bata document load garna
from langchain_community.vectorstores import Chroma  #  Vector database store garna

from langchain_core.output_parsers import StrOutputParser  #  Output parse garna
from langchain_core.runnables import RunnablePassthrough  #  Input pass-through garna

from langchain_huggingface import (
    HuggingFaceEmbeddings,
)  #  HuggingFace embeddings generate garna
from langchain_ollama import ChatOllama  #  Local Ollama LLM use garna
from langchain_core.prompts import ChatPromptTemplate  #  Prompt template banau na

# INDEXING


# 1.  Load Documents
loader = WebBaseLoader(
    web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("post-content", "post-title", "post-header")
        )
    ),
)
docs = loader.load()  #  Website bata document load garna


# 2.  Split Documents
# text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
# testing other tokenizer BPE
from langchain_text_splitters import TokenTextSplitter

text_splitter = TokenTextSplitter(
    chunk_size=256,
    chunk_overlap=50,
)
splits = text_splitter.split_documents(docs)


# 3.  Create Embeddings
embeddings = HuggingFaceEmbeddings(model_name="BAAI/bge-small-en-v1.5")


# 4.  Store Embeddings in Vector Database
vectorstore = Chroma.from_documents(documents=splits, embedding=embeddings)


# 5.  Create Retriever
retriever = vectorstore.as_retriever()


# RETRIEVAL + GENERATION


# 6.Load RAG Prompt
prompt = ChatPromptTemplate.from_messages(
    [("system", "You are a helpful assistant."), ("user", "{question}")]
)


# 7. Local LLM
llm = ChatOllama(model="llama3:latest", temperature=0)


# 8. Format Documents
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)


# 9. Create RAG Chain
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)


# ASK QUESTION


response = rag_chain.invoke("full form of HTML")  #  Question sodhne
print(response)  #  Answer print garne
```
